```python
# Question 1(b)
# extracting confirmed cases as my dependent variable from the dataset
y = df['Confirmed Cases']
y
```

```python
# extracting socioeconomic as my independent variable from the dataset
x = df['Normalized Deaths']
x = x[:, np.newaxis]
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting the data for training and testing
xtrain, x_test, y_train, y_test = train_test_split(x,y, test_size=1/3, random_state=0)
xtrain, x_test, y_train, y_test 
```

```python
# using the the linear regression class to fit
train_model = LinearRegression()
train_model
```

```python
# fitting
train_model.fit(xtrain,y_train)
```

```python
# predicting the results to get new output
y_predict = train_model.predict(x_test)
y_predict
```

```python
x_predict = train_model.predict(xtrain)
x_predict
```

```python
# visualizing the trained results
plt.scatter(xtrain, y_train, color="red")
```

```python
# visualizing the trained and predicted results
plt.scatter(xtrain, x_predict, color="blue")
```

```python
# visualising my trained values alongside the predicted values
plt.scatter(xtrain, y_train, color="red")
plt.scatter(xtrain, x_predict, color="blue")
plt.title("A scatter plot showing confirmed cases alongside normalised death")
plt.xlabel("Normalised cases")
plt.ylabel("Confirmed cases")
```

```python
# visualising my tested values alongside the predicted values
plt.scatter(x_test, y_test, color="red")
plt.scatter(xtrain, x_predict, color="blue")
plt.title("A scatter plot showing confirmed cases alongside normalised death")
plt.xlabel("Normalised cases")
plt.ylabel("Confirmed cases")
```
